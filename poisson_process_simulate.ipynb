# poisson_process_simulate

```python
# import package
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1

```python
# Q1
# set seed
np.random.seed(411078042)

# set fig
sns.set(style="whitegrid")
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i in range(3):
    # set lambda
    lam = 1.5
    current_time = 0
    event_times = []
    event_count = 0
    event_counts = []

    while current_time < 10:
        current_time += np.random.exponential(1 / lam)
        event_count += 1
        event_times.append(current_time)
        event_counts.append(event_count)

    # plot
    axs[i].step(event_times, event_counts, where='post')
    axs[i].scatter(event_times, event_counts, c='black')
    axs[i].set_xlim(0, 10)
    axs[i].set_xlabel('time (t)')
    axs[i].set_ylabel('event count')
    axs[i].set_title(f'poisson process Model {i + 1}')

# show
plt.tight_layout()
plt.show()
```

## Q2

```python
# Q2
# set seed
np.random.seed(411078042)

# basic setting
loop_num = 1000
lam = 1.5
sum_event_counts = []

# 1000 times Sojourn
event_counts_list = []
for _ in range(loop_num):
    current_time = 0
    event_count = 0
    while current_time < 10:
        current_time += np.random.exponential(1 / lam)
        event_count += 1
    event_counts_list.append(event_count - 1)

# set fig
sns.set(style="whitegrid")
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# set bins
bins = np.arange(0, max(event_counts_list) + 2) - 0.5

# Sojourn X(10)
axs[0].hist(event_counts_list, bins=bins, density=True, alpha=0.75, color='skyblue')
axs[0].set_title('Sojourn X(10)')
axs[0].set_xlabel('count')
axs[0].set_ylabel('frequency')

# Poisson(15)
poisson_samples = np.random.poisson(15, loop_num)
bins_poisson = np.arange(0, max(poisson_samples) + 2) - 0.5
axs[1].hist(poisson_samples, bins=bins_poisson, density=True, alpha=0.75, color='maroon')
axs[1].set_title('Poisson(15)')
axs[1].set_xlabel('count')
axs[1].set_ylabel('frequency')


plt.tight_layout()
plt.show()
```
